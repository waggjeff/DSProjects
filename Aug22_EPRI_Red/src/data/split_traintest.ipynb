## Split the test and training data sets

This script takes the cleaned flow data and generates training (~80%) and test (~20%) datasets based on splitting the data by days. One can decide which days will be used for testing, while the rest will be used for training (J Wagg). 

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
flow_data = pd.read_csv("final_5_1.csv") # added sep=';' for Alek's latest file

flow_data['index'] = pd.to_datetime(flow_data['index'])
print(flow_data['index'])
```

```python
flow_test = pd.DataFrame()
flow_train = pd.DataFrame()

# This is where you define the dates used for testing. Note that we only have two days with cyber attacks, 
# so we split these between the test and train samples
for index, row in flow_data.iterrows():
    date = row['index']
    if (date.month == 7 and date.day == 19) or (date.month == 7 and date.day == 27) or (date.month == 8 and date.day == 1):
        flow_test = flow_test.append([ row ])
    else:
        flow_train = flow_train.append( [row] )
```

```python
flow_test.to_csv("test_wzeek_v2.csv", encoding='utf-8', index=False)
flow_train.to_csv("train_wzeek_v2.csv", encoding='utf-8', index=False)
```
