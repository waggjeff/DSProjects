```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
from matplotlib import pyplot as plt
import numpy as np
warnings.filterwarnings('ignore')
from datetime import timedelta
```

```python
import flow_loop_preprocess_XW as process
```

```python
path="../../data/Rockwell_Flow_Loop_Data/Flow_Loop_Data_072022.csv"
```

```python
data=pd.read_csv(path)
```

```python
data.head(5)
```

We preprocess the data first:

```python
data=process.table_formattting(data)
```

```python
data.columns
```

```python
data.head(5)
```

Transform to uniform timestamp:

```python
for col in data:
    if data[col].dtype == 'O':
        data[col] = pd.to_datetime(data[col])

i = 0
while i <= (data.shape[1] - 2):
    sub_df = process.pull_feature(i, i+1, data)
    sub_df = sub_df.resample('10S').mean()
    if i == 0:
        df = sub_df
    else:
        df = df.join(sub_df, how='outer')
    i += 2
data = df.loc[:, df.columns.str.contains('Raw')]

data.head(5)
```

```python
fig, ax=plt.subplots(1,1,figsize=(20,6))
sns.lineplot(x=data.index,y="Loop 1 Flow Raw Value",data=data,label="Loop 1 Flow")
sns.lineplot(x=data.index,y="Loop 2 Flow Raw Value",data=data,label="Loop 2 Flow")
```

Next, we plot an isolated anomaly.

```python
#Xifan`s plotting script:
def anomalie_detection(begin, end, event_name):
    event_begin = pd.to_datetime('2022-07-20 ' + begin)
    event_end = pd.to_datetime('2022-07-20 ' + end)
    pre_time = event_begin - timedelta(minutes=2)
    aft_time = event_end + timedelta(minutes=2)

    sub_df = data[pre_time : aft_time]
    fig, ax = plt.subplots(figsize=(18,8))
    ax.plot(sub_df,'o-', label=data.columns)
    ax.vlines([event_begin, event_end], 2, 18000, colors='black', linestyles='dashed' )
    ax.set_yscale('log')
    ax.legend()
    ax.set_title(f'{event_name}, begin: {begin}, end: {end}', fontsize=14)
    plt.show()
```

```python
event_dic = {
    'SV_102_A_air_close_open': ['10:48:00', '10:51:00'],
    'SV_202_A_air_close_open': ['11:04:00', '11:07:00'],
    'SV_102_B_air_close_open': ['11:28:00', '11:31:00'],
    'SV_202_B_air_close_open': ['11:41:00', '11:42:00'],
    "LCV_203_air_close_open": ['12:19:00', '12:23:00'],
    "LCV_103_air_close_open": ['12:47:00', '12:57:00'],
     'SV_201_air_close_open': ['14:51:00', '14:56:00'],
    'SV_101_air_close_open': ['15:05:00', '15:08:00'],
    'overall_instrument_air': ['15:40:00', '15:40:00'],
}
```
