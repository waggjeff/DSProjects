```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="../../data/Rockwell_Flow_Loop_Data/Flow_Loop_Data_072022.csv"
```

```python
data=pd.read_csv(path)
```

```python
data.head(10)
```

```python
data.info()
```

From the table above, we see that the "Description" is in the fifth line. This line seems to contain the features.

```python
data.iloc[4,:]
```

```python
data.iloc[7,:]
```

Here, we see that we always have the timestamp of the observation, followed by the numerical value of the features. First, we delete the lines at the top that contain no numerical data:

```python
data=data.iloc[7:,:]
data.head(5)
```

Now, the rename the features and only use the raw values of the features going forward:

```python
data["Loop 1 Flow Raw Time"]=data["Stop time"]
data["Loop 1 Flow Raw"]=data["7/20/22 8pm"]
data2=data[["Loop 1 Flow Raw Time","Loop 1 Flow Raw"]]
data2["Loop 2 Flow Raw Time"]=data["Unnamed: 7"]
data2["Loop 2 Flow Raw"]=data["Unnamed: 8"]
data2["Tank 1 Level Raw Value Time"]=data["Unnamed: 11"]
data2["Tank 1 Level Raw Value"]=data["Unnamed: 12"]
data2["Tank 2 Level Raw Value Time"]=data["Unnamed: 15"]
data2["Tank 2 Level Raw Value"]=data["Unnamed: 16"]
data2["Pump 1 Inlet Pressure Raw Value Time"]=data["Unnamed: 19"]
data2["Pump 1 Inlet Pressure Raw Value"]=data["Unnamed: 20"]
data2["Pump 1 Outlet Pressure Raw Value Time"]=data["Unnamed: 23"]
data2["Pump 1 Outlet Pressure Raw Value"]=data["Unnamed: 24"]
data2["Pump 2 Inlet Pressure Raw Value Time"]=data["Unnamed: 27"]
data2["Pump 2 Inlet Pressure Raw Value"]=data["Unnamed: 28"]
data2["Pump 2 Outlet Pressure Raw Value Time"]=data["Unnamed: 31"]
data2["Pump 2 Outlet Pressure Raw Value"]=data["Unnamed: 32"]
data2["Pump 1 Speed Feedback Raw Value Time"]=data["Unnamed: 35"]
data2["Pump 1 Speed Feedback Raw Value"]=data["Unnamed: 36"]
data2["Pump 2 Speed Feedback Raw Value Time"]=data["Unnamed: 39"]
data2["Pump 2 Speed Feedback Raw Value"]=data["Unnamed: 40"]
```

```python
data2.head(5)
```

```python
data2.dtypes
```

We see here that all datatypes are at the moment objects. Now we replace all empty entries with nan values. 

```python
data2.replace(' ', np.nan, inplace=True)
```

```python
data2["Loop 1 Flow Raw Time"]=pd.to_datetime(data2["Loop 1 Flow Raw Time"],errors="coerce")
data2["Loop 2 Flow Raw Time"]=pd.to_datetime(data2["Loop 2 Flow Raw Time"],errors="coerce")
data2["Tank 1 Level Raw Value Time"]=pd.to_datetime(data2["Tank 1 Level Raw Value Time"],errors="coerce")
data2["Tank 2 Level Raw Value Time"]=pd.to_datetime(data2["Tank 2 Level Raw Value Time"],errors="coerce")
data2["Pump 1 Inlet Pressure Raw Value Time "]=pd.to_datetime(data2["Pump 1 Inlet Pressure Raw Value Time"],errors="coerce")
data2["Pump 2 Inlet Pressure Raw Value Time "]=pd.to_datetime(data2["Pump 2 Inlet Pressure Raw Value Time"],errors="coerce")
data2["Pump 1 Outlet Pressure Raw Value Time"]=pd.to_datetime(data2["Pump 1 Outlet Pressure Raw Value Time"],errors="coerce")
data2["Pump 2 Outlet Pressure Raw Value Time"]=pd.to_datetime(data2["Pump 2 Outlet Pressure Raw Value Time"],errors="coerce")
data2["Pump 1 Speed Feedback Raw Value Time"]=pd.to_datetime(data2["Pump 1 Speed Feedback Raw Value Time"],errors="coerce")
data2["Pump 2 Speed Feedback Raw Value Time"]=pd.to_datetime(data2["Pump 2 Speed Feedback Raw Value Time"],errors="coerce")
```

```python
data2["Loop 1 Flow Raw"]=pd.to_numeric(data2["Loop 1 Flow Raw"],errors="coerce")
data2["Loop 2 Flow Raw"]=pd.to_numeric(data2["Loop 2 Flow Raw"],errors="coerce")
data2["Tank 1 Level Raw Value"]=pd.to_numeric(data2["Tank 1 Level Raw Value"],errors="coerce")
data2["Tank 2 Level Raw Value"]=pd.to_numeric(data2["Tank 2 Level Raw Value"],errors="coerce")
data2["Pump 1 Inlet Pressure Raw Value"]=pd.to_numeric(data2["Pump 1 Inlet Pressure Raw Value"],errors="coerce")
data2["Pump 2 Inlet Pressure Raw Value"]=pd.to_numeric(data2["Pump 2 Inlet Pressure Raw Value"],errors="coerce")
data2["Pump 1 Outlet Pressure Raw Value"]=pd.to_numeric(data2["Pump 1 Outlet Pressure Raw Value"],errors="coerce")
data2["Pump 2 Outlet Pressure Raw Value"]=pd.to_numeric(data2["Pump 2 Outlet Pressure Raw Value"],errors="coerce")
data2["Pump 1 Speed Feedback Raw Value"]=pd.to_numeric(data2["Pump 1 Speed Feedback Raw Value"],errors="coerce")
data2["Pump 2 Speed Feedback Raw Value"]=pd.to_numeric(data2["Pump 2 Speed Feedback Raw Value"],errors="coerce")
```
