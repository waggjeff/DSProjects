```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
from matplotlib import pyplot as plt
import numpy as np
warnings.filterwarnings('ignore')
from datetime import timedelta
```

```python
path="../../data/Rockwell_Flow_Loop_Data/final.csv"
data=pd.read_csv(path)
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
data["target"].value_counts()
```

```python
data["anomaly"]=data["target"]!="normal"
```

```python
y_train=data["anomaly"]
features=data
features.drop(["anomaly","target"],axis=1)
corrs=features.corrwith(y_train)
corrs.sort_values(ascending=False,key=abs)
```

```python
fig, ax=plt.subplots(1,1,figsize=(12,8))
features2=data.drop(["anomaly"],axis=1)
sns.heatmap(data=data.corr())
plt.title("Correlation heatmap of features and anomalies")
plt.savefig("correlation_features_all.pdf",bbox_inches='tight')
plt.show()
```

Let`s zoom into the gradients for some of the features to see how much they are correlated.

```python
fig, ax=plt.subplots(1,1,figsize=(12,8))
features_gradient=data[["max_FIT102_Raw_Signal","min_FIT102_Raw_Signal","mean_FIT102_Raw_Signal","std_FIT102_Raw_Signal"]]
sns.heatmap(data=features_gradient.corr())
plt.title("Correlation heatmap of feature gradients")
plt.show()
```
