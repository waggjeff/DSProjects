```python
raw = pd.read_csv('./Data/Variable Data.csv', header=None)
```

```python
raw.iloc[5,:]
```

```python
def eda(x, y):
    data = raw.iloc[8:,[x, y]]
    data.replace(' ', np.nan, inplace=True)
    data = data.dropna()
    data = data.iloc[:-1, :]
    data[x] = pd.to_datetime(data[x], errors='coerce')
    data[y] = data[y].astype('float')
    data.set_index([x], inplace=True)
    #print(data.head(20))
    title = raw.iloc[5, y]
    plt.figure(figsize=(20, 6))
    plt.plot(data)
    plt.title(title)
```

```python
i = 2
while i <= 40:
    eda(i, i+1)
    i += 4
```
